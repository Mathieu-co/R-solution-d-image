```python
import pandas as pd
import matplotlib.pyplot as plt

# Chargement des données depuis les fichiers CSV
srf_2_data = pd.read_csv('statistics/srf_2_train_results.csv')
srf_4_data = pd.read_csv('statistics/srf_4_train_results.csv')
srf_8_data = pd.read_csv('statistics/srf_8_train_results.csv')
```

```python
def plot_ssim(data, model_name):
    # Création du graphique
    plt.figure(figsize=(10, 6))

    # Tracer les valeurs SSIM pour chaque modèle
    plt.plot(data['Epoch'], data['SSIM'], label='Validation SSIM', marker='o')
    plt.plot(data['Epoch'], data['Train_SSIM'], label='Train SSIM', marker='o')

    # Ajout de titres et labels
    plt.title('SSIM de '+ model_name + ' en fonction époques')
    plt.xlabel('Époque')
    plt.ylabel('SSIM')
    plt.legend()
    plt.grid(True)

    # Affichage du graphique
    plt.show()

plot_ssim(srf_2_data, 'facteur 2')
plot_ssim(srf_4_data, 'facteur 4')
plot_ssim(srf_8_data, 'facteur 8')
```

```python
def plot_psnr(data, model_name):
    # Création du graphique
    plt.figure(figsize=(10, 6))

    # Tracer les valeurs SSIM pour chaque modèle
    plt.plot(data['Epoch'], data['PSNR'], label='Validation PSNR', marker='o')
    plt.plot(data['Epoch'], data['Train_PSNR'], label='Train PSNR', marker='o')

    # Ajout de titres et labels
    plt.title('PSNR de '+ model_name + ' en fonction époques')
    plt.xlabel('Époque')
    plt.ylabel('SSIM')
    plt.legend()
    plt.grid(True)

    # Affichage du graphique
    plt.show()

plot_psnr(srf_2_data, 'facteur 2')
plot_psnr(srf_4_data, 'facteur 4')
plot_psnr(srf_8_data, 'facteur 8')
```

```python
def plot_perf(data, model_name):

    plt.figure(figsize=(10, 6))

    plt.plot(data['Epoch'], data['Score_D'], label='Score_D', marker='o', color='green')
    plt.plot(data['Epoch'], data['Score_G'], label='Score_G', marker='s', color='purple')
    plt.title('Score du modèle ' + model_name + ' en fonction des Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

plot_perf(srf_2_data,'facteur 2')
plot_perf(srf_4_data,'facteur 4' )
plot_perf(srf_8_data,'facteur 8')
```

```python
def plot_loss_d(srf_2_data,srf_4_data,srf_8_data):
    plt.figure(figsize=(10, 6))

    plt.plot(srf_2_data['Epoch'], srf_2_data['Loss_D'], label='SRF 2', marker='o')
    plt.plot(srf_4_data['Epoch'], srf_4_data['Loss_D'], label='SRF 4', marker='s')
    plt.plot(srf_8_data['Epoch'], srf_8_data['Loss_D'], label='SRF 8', marker='^')

    plt.title('Perte du discriminateur en fonction des époques pour différents modèles')
    plt.xlabel('Époque')
    plt.ylabel('Loss_D')
    plt.legend()
    plt.grid(True)
    
plot_loss_d(srf_2_data,srf_4_data,srf_8_data)
```

```python
def plot_loss_g(srf_2_data,srf_4_data,srf_8_data):
    plt.figure(figsize=(10, 6))

    plt.plot(srf_2_data['Epoch'], srf_2_data['Loss_G'], label='SRF 2', marker='o')
    plt.plot(srf_4_data['Epoch'], srf_4_data['Loss_G'], label='SRF 4', marker='s')
    plt.plot(srf_8_data['Epoch'], srf_8_data['Loss_G'], label='SRF 8', marker='^')

    plt.title('Perte du générateur en fonction des époques pour différents modèles')
    plt.xlabel('Époque')
    plt.ylabel('Loss_G')
    plt.legend()
    plt.grid(True)
    
plot_loss_g(srf_2_data,srf_4_data,srf_8_data)
```
